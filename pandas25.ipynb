### 25. Calcular el promedio de rating por tipo de app (paid/free) para las aplicaciones que posean al menos una review donde el modulo de la polaridad (positiva o negativa) sea superior a dos veces la media del modulo de la polaridad de su tipo. Aclaracion Una app con polaridad -0.5 tiene modulo 0.5, asi que si todas las apps pagas tienen 0.5 o -0.5 en polaridad, el promedio del modulo de la polaridad es 0.5 (no 0) ⭐⭐

```python
import pandas as pd
```

```python
apps = pd.read_csv('googleplaystore.csv')
reviews = pd.read_csv('googleplaystore_user_reviews.csv')
```

Voy a necesitar, de apps, las columnas "App", "Rating" y "Type"; y de reviews "App" y "Sentiment_Polarity"; me quedo con esas

```python
apps = apps.drop_duplicates()[['App', 'Rating', 'Type']]
reviews = reviews[['App', 'Sentiment_Polarity']]
```

Creo un nuevo dataframe, que surge del inner merge de los anteriores, y le agrego el modulo de la polaridad y la media del tipo

```python
df = pd.merge(apps, reviews, how='inner')
df['modulo_polaridad'] = abs(df.Sentiment_Polarity)
df['media_tipo'] = df.groupby('Type')['Sentiment_Polarity'].transform('mean')
```

Me quedo con las apps cuyo módulo de la polaridad sea mayor al doble de la media de su tipo

```python
df = df[df.modulo_polaridad > 2 * df.media_tipo]
df
```

```python
df.drop_duplicates(subset='App').groupby('Type').Rating.mean().reset_index()
```
