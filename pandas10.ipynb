### 10. Mostrar el largo promedio de las reviews por tipo de sentimiento ⭐

```python
import pandas as pd
```

```python
reviews = pd.read_csv('googleplaystore_user_reviews.csv')
reviews
```

Voy a limpiar las reviews con NaN en la columna "Translated_Review"

```python
reviews.dropna(subset='Translated_Review', inplace=True)
```

```python
reviews['largo_review'] = reviews.Translated_Review.map(lambda x: len(x))
reviews = reviews[['Sentiment', 'largo_review']]
reviews.groupby('Sentiment').mean()
```
